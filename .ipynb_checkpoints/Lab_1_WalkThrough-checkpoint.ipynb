# Linear Regression on Boston Housing Dataset

This Housing dataset contains information about different houses in Boston. This data was originally a part of UCI Machine Learning Repository and has been removed now.

There are 506 samples and 13 feature variables in this dataset. The objective is to predict the value of prices of the house using the given features. 
The dataset itself is available at https://archive.ics.uci.edu/ml/datasets/Housing . However, because we are going to use scikit-learn, we can import it right away from the scikit-learn itself.

First, we will import the required libraries.

```python
import numpy as np
import pandas as pd

#Visualization Libraries
import seaborn as sns
import matplotlib.pyplot as plt

#To plot the graph embedded in the notebook
%matplotlib inline
```

```python
#imports from sklearn library

from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
#loading the dataset direclty from sklearn
boston = datasets.load_boston()
```

sklearn returns Dictionary-like object, the interesting attributes are: ‘data’, the data to learn, ‘target’, the regression targets, ‘DESCR’, the full description of the dataset, and ‘filename’, the physical location of boston csv dataset. This we can from the following Operations 

```python
print(type(boston))
print('\n')
print(boston.keys())
print('\n')
print(boston.data.shape)
print('\n')
print(boston.feature_names)
```

```python
print(boston.DESCR)
```

```python
#convert the dataset into a pandas dataframe
bos = pd.DataFrame(boston.data, columns = boston.feature_names)

print(type(bos))
print('\n')
print(bos.shape)
print('\n')
print(bos.head())
```

```python
print(boston.target.shape)
```

```python
#insert the 'target' column in main dataframe
bos['PRICE'] = boston.target
print(bos.head())
```

## Exploratory Data Analysis

Exploratory Data Analysis is a very important step before training the model. Here, we will use visualizations to understand the relationship of the target variable with other features.

Let’s first plot the distribution of the target variable $ Price $. We will use the distplot function from the seaborn library.

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.hist(bos['PRICE'], bins=30)
plt.xlabel("House prices in $1000")
plt.show()
```

We can see from the plot that the values of **PRICE** are distributed normally with few outliers. Most of the house are around 20-24 range (in $1000 scale)

Now, we create a correlation matrix that measures the linear relationships between the variables. The correlation matrix can be formed by using the corr function from the pandas dataframe library. We will use the heatmap function from the seaborn library to plot the correlation matrix.

```python
#Created a correlation matrix
correlation_matrix = bos.corr().round(2)
# annot = True to print the values inside the square
sns.heatmap(data=correlation_matrix, annot=True)
```

The correlation coefficient ranges from -1 to 1. If the value is close to 1, it means that there is a strong positive correlation between the two variables. When it is close to -1, the variables have a strong negative correlation.

### Notice

1. By looking at the correlation matrix we can see that *RM* has a strong positive correlation with *PRICE* **(0.7)** where as *LSTAT* has a high negative correlation with *PRICE* **(-0.74)**.
2. An important point in selecting features for a linear regression model is to check for multicolinearity. The features *RAD*, *TAX* have a correlation of **0.91**. These feature pairs are strongly correlated to each other. This can affect the model. Same goes for the features *DIS* and *AGE* which have a correlation of *-0.75*.

But for now we will keep all the features.

```python
plt.figure(figsize=(20, 5))

features = ['LSTAT', 'RM']
target = bos['PRICE']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = bos[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title("Variation in House prices")
    plt.xlabel(col)
    plt.ylabel('"House prices in $1000"')
```

### Notice
1. The prices increase as the value of RM increases linearly. There are few outliers and the data seems to be capped at 50.
2. The prices tend to decrease with an increase in LSTAT. Though it doesn’t look to be following exactly a linear line.

Since it is really hard to visualise with the multiple features, we will first predict the house price with just one vaiable and then move to the regression with all features.

### Univariate regression model
You saw that some variables are highly correlated with the house prices in Boston. Let's make the univariate regression model with the variables have the **absolute correlation coefficient value over 0.7.** (In this case, RM and LSTAT satisfied the condition so makes two different models using each of them.)


```python
# created np array with target variable
y_price = bos.PRICE
y_price = np.array(y_price).reshape(-1,1)

print(y_price.shape)

#####################
## create np arrays for RM and LSTAT (You make two different arrays for both of them)
## type your own code

X_rooms = bos.RM
X_rooms = np.array(X_rooms).reshape(-1,1)
X_lstat = bos.LSTAT
X_lstat = np.array(X_lstat).reshape(-1,1)


#####################

print(X_rooms.shape)
print(X_lstat.shape)
```

### Splitting the data into training and testing sets

SInce we need to test our model, we split the data into training and testing sets. We train the model with 80% of the samples and test with the remaining 20%. We do this to assess the model’s performance on unseen data. 

To split the data we use train_test_split function provided by scikit-learn library. We finally print the shapes of our *training* and *test* set to verify if the splitting has occurred properly.

```python
#####################
## make train and test datasets (e.g. first dataset using 'RM' and 'PRICE', second dataset using 'LSTAT' and 'PRICE')
## type your own code using 'train_test_split' function
## random_state = 5
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_rooms, y_price, train_size=0.80, random_state=5)

X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_lstat, y_price, train_size=0.80, random_state=5)
#####################
```

```python
## check the data shape of train and test dataset
print(X_train_1.shape)
print(X_test_1.shape)
print(y_train_1.shape)
print(y_test_1.shape)
print('\n')
print(X_train_2.shape)
print(X_test_2.shape)
print(y_train_2.shape)
print(y_test_2.shape)
```

### Training and testing the model
Here we use scikit-learn’s LinearRegression to train our model on both the training and check it on the test sets

```python
#####################
## Make linear regression models and train the models (reg_1 with RM, reg_2 with LSTAT and fit the model with train data)
## Type your own code
reg_1 = LinearRegression().fit(X_train_1, y_train_1)
reg_2 = LinearRegression().fit(X_train_2, y_train_2)
#####################
```

```python
#####################
## model evaluation for training set
# (1) predict house price with train data
# (2) evaluate the results with RMSE(root mean square error) and r^2 score
#   RMSE = np.sqrt(mean_squared_error(y_trueValue, y_predictedValue))
## Type you own code

predict_1 = reg_1.predict(X_train_1)
rmse_1 = np.sqrt(mean_squared_error(y_train_1, predict_1))
r2_1 = r2_score(y_train_1, predict_1)


predict_2 = reg_2.predict(X_train_2)
rmse_2 = np.sqrt(mean_squared_error(y_train_2, predict_2))
r2_2 = r2_score(y_train_2, predict_2)

#####################
```

```python
print("MODEL 1 with RM")
print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse_1))
print('R2 score is {}'.format(r2_1))
print("\n")

print("MODEL 2 with LSTAT")
print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse_2))
print('R2 score is {}'.format(r2_2))
print("\n")
```

```python
#####################
# model evaluation for test set
# (1) predict house price with test data (use 'predict' with test data)
# (2) evaluate the results with RMSE(root mean square error) and r^2 score
#   RMSE = np.sqrt(mean_squared_error(y_trueValue, y_predictedValue))

## Type you own code
predict_1 = reg_1.predict(X_test_1)
rmse_1 = np.sqrt(mean_squared_error(y_test_1, predict_1))
r2_1 = r2_score(y_test_1, predict_1)


predict_2 = reg_2.predict(X_test_2)
rmse_2 = np.sqrt(mean_squared_error(y_test_2, predict_2))
r2_2 = r2_score(y_test_2, predict_2)

#####################
```

```python
print("MODEL 1 with RM")
print("The model performance for test set")
print("--------------------------------------")
print("Root Mean Squared Error: {}".format(rmse_1))
print("R^2: {}".format(r2_1))
print("\n")

print("MODEL 2 with LSTAT")
print("The model performance for test set")
print("--------------------------------------")
print("Root Mean Squared Error: {}".format(rmse_2))
print("R^2: {}".format(r2_2))
print("\n")
```

```python
prediction_space = np.linspace(min(X_rooms), max(X_rooms)).reshape(-1,1) 
plt.scatter(X_rooms,y_price)
plt.plot(prediction_space, reg_1.predict(prediction_space), color = 'black', linewidth = 3)
plt.ylabel('value of house/1000($)')
plt.xlabel('number of rooms')
plt.show()
```

```python
prediction_space = np.linspace(min(X_lstat), max(X_lstat)).reshape(-1,1) 
plt.scatter(X_lstat,y_price)
plt.plot(prediction_space, reg_2.predict(prediction_space), color = 'black', linewidth = 3)
plt.ylabel('value of house/1000($)')
plt.xlabel('% lower status of the population')
plt.show()
```

### Regression Model for All the variables

Now we will create a model considering all the features in the dataset. The process is almost the same and also the evaluation model but in this case the visualization will not be possible in a 2D space.

```python
X = bos.drop('PRICE', axis = 1)
y = bos['PRICE']
```

```python
#####################
## make train and test datasets 
## type your own code using train_test_split function
## test_size = .2, random_state = 42


print(X.shape)
print(y.shape)
print('\n')
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.80, random_state=42)

print(X_train.shape)
print(y_train.shape)
print('\n')
print(X_test.shape)
print(y_test.shape)

#####################

```

```python
#####################
## Make linear regression models and train the models (reg_all)
## Type your own code

reg_all = LinearRegression().fit(X_train, y_train)

#####################
```

```python
#####################
## model evaluation for training set
# (1) predict house price with train data
# (2) evaluate the results with RMSE(root mean square error) and r^2 score
#   RMSE = np.sqrt(mean_squared_error(y_trueValue, y_predictedValue))
## Type you own code

predict = reg_all.predict(X_train)
rmse = np.sqrt(mean_squared_error(y_train, predict))
r2 = r2_score(y_train, predict)



#####################

```

```python
print("The model performance for training set")
print("--------------------------------------")
print('Root Mean Squared Error: {}'.format(rmse))
print('R^2: {}'.format(r2))
print("\n")
```

```python
#####################
# model evaluation for test set
# (1) predict house price with test data
# (2) evaluate the results with RMSE(root mean square error) and r^2 score
#   RMSE = np.sqrt(mean_squared_error(y_trueValue, y_predictedValue))

## Type you own code


predict = reg_all.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, predict))
r2 = r2_score(y_test, predict)





#####################
```

```python
print("The model performance for test set")
print("--------------------------------------")
print("Root Mean Squared Error: {}".format(rmse))
print("R^2: {}".format(r2))
print("\n")
```

```python
names = boston.feature_names
reg_all_coef = reg_all.fit(X_train, y_train).coef_

coef_df = pd.DataFrame(names, columns = ['Variables'])
coef_df['Coef'] = reg_all_coef

print(coef_df)

plt.bar(range(len(names)),reg_all_coef)
plt.xticks(range(len(names)),names, rotation = 60)
plt.ylabel('Coefficient')
plt.show()
```

### Lasso and Ridge Regression Models for All the variables

Now we will create a lasso and ridge regression models with the same dataset including all variables. Lasso and Ridge regression models are widely used for feature selection. As we learn from the lecture, they use different penalty function (Lasso: L1 norm, Ridge: L2 norm). 

```python
#####################
# (1) Make Ridge regression models and train the models (alpha = .1)
# (2) predict house price with test data
# (3) evaluate the results with RMSE(root mean square error) and r^2 score
#   RMSE = np.sqrt(mean_squared_error(y_trueValue, y_predictedValue))
## Type your own code

ridge = Ridge(alpha=0.1).fit(X_train, y_train)
predict_ridge = ridge.predict(X_test)
rmse_ridge = np.sqrt(mean_squared_error(y_test, predict_ridge))
r2_ridge = r2_score(y_test, predict_ridge)


#####################


```

```python
print("Ridge Regreesion Model")
print("The model performance for test set")
print("--------------------------------------")
print("Root Mean Squared Error: {}".format(rmse_ridge))
print("R^2: {}".format(r2_ridge))
print("\n")
```

```python
names = boston.feature_names
ridge_all_coef = ridge.fit(X_train, y_train).coef_

coef_df = pd.DataFrame(names, columns = ['Variables'])
coef_df['Coef'] = ridge_all_coef

print(coef_df)

plt.bar(range(len(names)),ridge_all_coef)
plt.xticks(range(len(names)),names, rotation = 60)
plt.ylabel('Coefficient')
plt.show()
```

```python
#####################
# (1) Make Lasso regression models and train the models (alpha = .1)
# (2) predict house price with test data
# (3) evaluate the results with RMSE(root mean square error) and r^2 score
#   RMSE = np.sqrt(mean_squared_error(y_trueValue, y_predictedValue))

## Type your own code


lasso = Lasso(alpha=0.1).fit(X_train, y_train)
predict_lasso = lasso.predict(X_test)
rmse_lasso = np.sqrt(mean_squared_error(y_test, predict_lasso))
r2_lasso = r2_score(y_test, predict_lasso)



#####################

```

```python
print("Lasso Regreesion Model")
print("The model performance for test set")
print("--------------------------------------")
print("Root Mean Squared Error: {}".format(rmse_lasso))
print("R^2: {}".format(r2_lasso))
print("\n")
```

```python
names = boston.feature_names
lasso_all_coef = lasso.fit(X_train, y_train).coef_

coef_df = pd.DataFrame(names, columns = ['Variables'])
coef_df['Coef'] = lasso_all_coef

print(coef_df)

plt.bar(range(len(names)),lasso_all_coef)
plt.xticks(range(len(names)),names, rotation = 60)
plt.ylabel('Coefficient')
plt.show()
```
